```python
from PIL import Image
import matplotlib.pyplot as plt
import os

# Load the images
image_paths = os.listdir('./phyloplots/')# Add your image file paths here
images = [Image.open(os.path.join('phyloplots/',path)) for path in image_paths if '.png' in path]

# Constants for the grid
images_per_row = 5
padding = 10  # Space for the label

# Determine the size of each row and overall image
widths, heights = zip(*(i.size for i in images))
max_width = max(widths) * images_per_row
total_rows = len(images) // images_per_row + (1 if len(images) % images_per_row else 0)
total_height = sum(max(heights[i * images_per_row:(i + 1) * images_per_row]) + padding for i in range(total_rows))

# Create a new image for the combined grid
combined_image = Image.new('RGB', (max_width, total_height), 'white')

# Paste the images onto the grid
y_offset = 0
for row in range(total_rows):
    x_offset = 0
    for col in range(images_per_row):
        index = row * images_per_row + col
        if index < len(images):
            img = images[index]
            combined_image.paste(img, (x_offset, y_offset + padding))
            x_offset += img.width
    row_height = max(heights[row * images_per_row:(row + 1) * images_per_row])
    y_offset += row_height + padding

# Save the combined image temporarily to add labels with matplotlib
#combined_image.save('combined_temp.jpg')

# Add labels using matplotlib
fig, ax = plt.subplots(figsize=(12, 12))
ax.imshow(combined_image)
ax.axis('off')  # Hide axis

# Add labels
for i, img in enumerate(images):
    row = i // images_per_row
    col = i % images_per_row
    x_label = col * max_width // images_per_row + 10
    y_label = row * (img.height + padding) + 5
    ax.text(x_label, y_label, f'Position {i+1}', fontsize=10, color='black')

# Save or display the final image
plt.savefig('combined_labeled.pdf', dpi=2000)
plt.show()
```
